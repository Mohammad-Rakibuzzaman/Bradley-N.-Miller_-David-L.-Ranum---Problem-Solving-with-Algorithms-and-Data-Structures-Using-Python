```python
squares = { x:x**2 for x in range(10)}

print(squares)
```

```python


for i in range(10):
    print(f"{i}: {i ** 2}", end=" , ")    
```

```python
squares = {}

for i in range(10):
    squares[i] = i ** 2

# Print the dictionary in the format "key: value" without the last comma
formatted_output = ', '.join([f"{key}: {value}" for key, value in squares.items()])

print(formatted_output)
```
